```python
import numpy as np
import pandas as pd

# Parameters
string_length = 30
population_size = 100
mutation_rate = 0.0001
crossover_rate = 0.95
max_generations = 100

# Initialize population (randomly)
population = np.random.choice([0, 1], size=(population_size, string_length))

def fitness(individual):
   
    return np.sum(individual)

def select_parent(population, fitnesses):
   #roulette wheel selection
    total_fitness = np.sum(fitnesses)
    pick = np.random.rand() * total_fitness
    current = 0
    for i, individual in enumerate(population):
        current += fitnesses[i]
        if current > pick:
            return individual

def crossover(parent1, parent2):
    #one-point crossover
    if np.random.rand() < crossover_rate:
        point = np.random.randint(1, string_length-1)
        child1 = np.concatenate((parent1[:point], parent2[point:]))
        child2 = np.concatenate((parent2[:point], parent1[point:]))
        return child1, child2
    else:
        return parent1, parent2

def mutate(individual):
    
    for i in range(string_length):
        if np.random.rand() < mutation_rate:
            individual[i] = 1 - individual[i]
    return individual

def evolve(population):
    
    new_population = []
    fitnesses = np.array([fitness(individual) for individual in population])
    for _ in range(population_size // 2):
        # Select parents
        parent1 = select_parent(population, fitnesses)
        parent2 = select_parent(population, fitnesses)
        # Crossover
        child1, child2 = crossover(parent1, parent2)
        # Mutate
        child1 = mutate(child1)
        child2 = mutate(child2)
        # Add to new population
        new_population.extend([child1, child2])
    return np.array(new_population)

# Store average fitness per generation
avg_fitness_history = []

# Evolution loop
for generation in range(max_generations):
    population = evolve(population)
    avg_fitness = np.mean([fitness(individual) for individual in population])
    avg_fitness_history.append(avg_fitness)
    print(f"Generation {generation}: Avg Fitness = {avg_fitness}")
```
